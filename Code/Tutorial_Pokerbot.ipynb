```python
env = make('no-limit-holdem', config={'game_num_players': 2})
state, player_id = env.reset()

done = False
while not done:
    print(state['raw_obs']['hand'])
    
    if player_id == 0:
        legal_actions = state['raw_legal_actions'][1]
        action = legal_actions
    else:
        legal_actions = state['raw_legal_actions'][2]
        action = legal_actions
    print(f"Player {player_id} chooses: {action} with hand {state['raw_obs']['hand']} and public cards {state['raw_obs']['public_cards']}")
    state, player_id = env.step(action)
    done = env.is_over()

# Get final payoffs
print("Payoffs:", env.get_payoffs())
```

```python
all_cards = ['H2', 'H5', 'H7', 'H9', 'HJ', 'D3', 'C8']
# Explanation:
# - 'H2', 'H5', 'H7', 'H9', 'HJ' are all hearts (so they form a flush),
# - The other two cards ('D3' for Diamonds and 'C8' for Clubs) are just extra cards.

# Create the Hand instance.
hand = Hand(all_cards)

# Evaluate the hand.
hand.evaluateHand()

# Output the evaluation results.
print("Evaluated Category:", hand.category)
print("Best Five Cards:", hand.get_hand_five_cards())
```
