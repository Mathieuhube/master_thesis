```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('./data.csv')
```

# Basic model: win or lose

Prediction made for (game, hand)

```python
X, y = data.loc[:, ~data.columns.isin(['won'])], data.loc[:, data.columns == 'won']
clf = LogisticRegression(random_state=1234).fit(X, y)
clf.score(X, y)
```
