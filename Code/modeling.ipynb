Let's do more complicated on **flop**. What do we need to pay attention to:

 - if there was a "fold" on pre-flop, we should remove such a data, it's useless
 - we can take all the features from pre-flop and add them to the data appeared after flop

```python
flop_only_features = [
 'player0_preflop_bet',
 'player0_preflop_check',
 'flop_hand_Flush',
 'flop_hand_Four of a Kind',
 'flop_hand_Full House',
 'flop_hand_High Card',
 'flop_hand_Pair',
 'flop_hand_Straight',
 'flop_hand_Three of a Kind',
 'flop_hand_Two Pair',
]

flop_features = flop_only_features + preflop_features

flop_target = 'player0_flop_fold'
```

```python
# Dropping unnecessary data with fold on pre-flop
flop_data = data[data[preflop_target] == 0]
# Making flop prediction
X, y = flop_data.loc[:, flop_data.columns.isin(flop_features)], flop_data.loc[:, flop_data.columns == flop_target]
clf = LogisticRegression(random_state=1234).fit(X, y)
clf.score(X, y)
```

```python
# Dropping unnecessary data with fold on flop
turn_data = flop_data[flop_data[flop_target] == 0]
# Making turn prediction
X, y = turn_data.loc[:, turn_data.columns.isin(turn_features)], turn_data.loc[:, turn_data.columns == turn_target]
clf = LogisticRegression(random_state=1234).fit(X, y)
clf.score(X, y)
```
