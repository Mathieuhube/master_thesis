```python
import gymnasium as gym
env = gym.make("FrozenLake-v1", render_mode="rgb_array")
env.reset()
for t in range(10):
   print("nTimestep {}".format(t))
   env.render()
   a = env.action_space.sample()
   ob, r, done, _ , info = env.step(a)
   print(a, ob, r, done, info)
   if done:
      print("nEpisode terminated early")
      break
```

```python
actions = {'Left': 0, 'Down': 1, 'Right': 2, 'Up': 3 }
good_plan = (2*['Down']) + ['Right'] + ['Down'] + (2*['Right'])
env = gym.make("FrozenLake-v1", is_slippery=False)
env.reset()
env.render()
for a in good_plan:
    new_state, reward, done, _ , info = env.step(actions[a])
    print(new_state)
    env.render()
    if done:
       break
```
